## Movies Data

```python
movies_credits_df = pd.read_csv('/content/drive/My Drive/CIS450 Project/Movies/movies_credits.csv')
movies_metadata_df = pd.read_csv('/content/drive/My Drive/CIS450 Project/Movies/movies_metadata.csv')
movies_ratings_df = pd.read_csv('/content/drive/My Drive/CIS450 Project/Movies/ratings.csv')
movies_links_df = pd.read_csv('/content/drive/My Drive/CIS450 Project/Movies/links.csv')
```

```python
from ast import literal_eval

movies_credits_df['cast'] = movies_credits_df['cast'].apply(literal_eval) #convert to list type
cast_df = movies_credits_df.explode('cast')
```

```python
movies_credits_df
```

```python
actors_df = pd.json_normalize(cast_df["cast"])
actors_df
```

```python
actors_df = actors_df[['id', 'name', 'gender']].dropna()
```

```python
actors_df = actors_df.rename(columns={'id':'Id', 'name': 'Name', 'gender' : 'Gender'}).drop_duplicates()
actors_df['Name'] = actors_df['Name'].str.replace(',','')
actors_df['Name'] = actors_df['Name'].str.replace('"','')
actors_df['Name'] = actors_df['Name'].str.replace('\'','')
```

```python
actors_df
```

```python
from google.colab import files

actors_df.to_csv('actors.csv',index=False)
```

```python
movies_credits_df['crew'] = movies_credits_df['crew'].apply(literal_eval) #convert to list type
crew_df = movies_credits_df.explode('crew')
```

```python
movie_directors_df = pd.json_normalize(crew_df["crew"])
```

```python
movie_directors_df = movie_directors_df[movie_directors_df['job'] == 'Director']
```

```python
directors_df = movie_directors_df[['id', 'name', 'gender']].dropna()
directors_df = directors_df.rename(columns={'id':'Id', 'name': 'Name', 'gender' : 'Gender'}).drop_duplicates()
```

```python
from google.colab import files

directors_df.to_csv('directors.csv',index=False)
```

```python
movie_crew_df = pd.json_normalize(crew_df["crew"])
list_of_movie_ids = list(crew_df['id'])
movie_crew_df['Movie_id'] = list_of_movie_ids
```

```python
movie_crew_df = movie_crew_df[movie_crew_df['job'] == 'Director']
movie_crew_df = movie_crew_df[['id', 'Movie_id']].dropna()
```

```python
crew_df.loc[crew_df['id'] == 5]
```

```python
directs_df = movie_crew_df.rename(columns={'id':'DirectorId', 'Movie_id': 'MovieId'}).drop_duplicates()
```

```python
directs_df.to_csv('directs.csv',index=False)
```

```python
movie_cast_df = pd.json_normalize(cast_df["cast"])
list_of_movie_ids = list(cast_df['id'])
movie_cast_df['Movie_id'] = list_of_movie_ids
```

```python
movie_cast_df = movie_cast_df[['cast_id', 'Movie_id', 'character']].dropna()
```

```python
plays_df = movie_cast_df.rename(columns={'cast_id':'ActorId', 'Movie_id': 'MovieId', 'character' : 'Character'}).drop_duplicates()
plays_df = plays_df[plays_df['Character'] != '']
```

```python
plays_df.to_csv('plays.csv',index=False)
```
