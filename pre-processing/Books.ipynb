### Authors

```python
import functools
allowed_chars = ['.','\'','-',' ','\t','\n']
foldr = lambda func, acc, xs: functools.reduce(lambda x, y: func(y, x), xs[::-1], acc)
def is_valid_author_name(name):
  return foldr(lambda c,acc: acc and (c in allowed_chars or c.isalnum() or c.isspace()), True, name)
is_valid_author_name('Kelly Milner Halls')
```

```python
# # Cleaning authors
authors = set(books_df['author'])
allowed_chars = ['.','\'','-']
cleaned_authors = set()
for author in authors:
  for name in author.split(','):
    name = name.strip()
    if is_valid_author_name(name) and name:
      cleaned_authors.add(name)
```

```python
author_df = pd.DataFrame(cleaned_authors)
author_df.columns = ['Name']
author_df
```
